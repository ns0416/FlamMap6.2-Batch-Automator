```python

"""
    FlamMap6.2 Batch Automator Verion 1.0.0
    ###
    Written By Nohsan Kim @ NAMISNT,
    School of Computer & Information Engineering in
    College of Software Convergence,
    Kwangwoon University SEOUL.
    2023-07-31
"""

from pywinauto import Application, keyboard
import psutil
import os
import time
import threading

PROCNAME = "FlamMap6.exe"

pids = []
threads = []
filepath = ""
runpath = ""
savepath = ""
timeout_unit = 5

GUI_IN_USE = False

DEVELOPER =  "\r\nFlamMap6.2 Batch Automator Verion 1.0.0\r\nDeveloped By Nohsan Kim @ NAMISNT, \r\nSchool of Computer & Information Engineering in \r\nCollege of Software Convergence,\r\nKwangwoon University SEOUL\r\n"

def searchPID():
    print("Flammap PID in Running : ")
    pid_exist = False
    for proc in psutil.process_iter():
        if proc.name() == PROCNAME:
            pid_exist = True
            print("PID : ", proc.pid)
    if pid_exist == False:
        print("Any Process that has name Flammap found")

def controlFlammap(pid, filepath, runpath, savepath, start_num, end_num, timeout_):
    print("Flammap PID : ", pid, "is going to Automation running")
    app = Application(backend='uia')
    app.connect(process=pid)
    dlg = app['FlamMap 6 - FlamMap1']
    #dlg.print_control_identifiers()
    open_landscape(dlg, timeout_)
    open_dialog(dlg, filepath, timeout_)
    for i in range(start_num, end_num+1):
        Wait_GUI_IN_USE()
        GUI_IN_USE = True
        import_run(dlg, timeout_)
        open_dialog(dlg, runpath+"\log"+str(i)+".txt", timeout_)
        run_window = get_run_window(dlg, timeout_)
        progress = launchMTT(run_window, timeout_)
        is_succ = isMTTFailed(app, progress)
        if is_succ == False:
            print("log"+str(i)+".txt : No Spread")
        else:
            continueProgress(progress)
        closeRunWindow(progress, run_window)
        selectTreeItem(dlg)
        if is_succ:
            exportRunToImg(dlg)
            save_dialog(dlg, savepath+"\log"+str(i)+".tif", timeout_)
            print("log"+str(i)+".txt : saved!")
        selectTreeItem(dlg)
        deleteRun(dlg)

def deleteRun(dlg):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        dlg.child_window(title="New Run", control_type="TreeItem").select()
    except:
        return
    try:
        keyboard.send_keys('{DELETE}')
        dlg['FlamMap'].wait('ready', timeout=5)
    except:
        GUI_IN_USE = False
        deleteRun(dlg)
    GUI_IN_USE = False
    clickDeleteBtn(dlg)
    

def clickDeleteBtn(dlg):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        dlg.child_window(title="예(Y)", auto_id="6", control_type="Button").click()
    except:
        GUI_IN_USE = False
        deleteRun(dlg)
        return
    GUI_IN_USE = False
    checkDeleteAlertClosed(dlg)

def checkDeleteAlertClosed(dlg):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        time.sleep(10)
        dlg.child_window(title="FlamMap", control_type="Window").wait('ready', timeout=2)
        GUI_IN_USE = False
        clickDeleteBtn(dlg)
    except:
        pass
    GUI_IN_USE = False

def exportRunToImg(dlg):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        keyboard.send_keys('{RIGHT}')
        dlg.child_window(title="MTT Arrival Time", control_type="TreeItem").select()
        dlg.child_window(title="MTT Arrival Time", control_type="TreeItem").click_input(button='right')
        keyboard.send_keys('{DOWN}')
        keyboard.send_keys('{ENTER}')
        dlg.child_window(title="다른 이름으로 저장", control_type="Window").wait('ready', timeout=10)
    except:
        GUI_IN_USE = False
        exportRunToImg(dlg)
    GUI_IN_USE = False
    

def selectTreeItem(dlg):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        dlg.child_window(title="New Run", control_type="TreeItem").select()
    except:
        GUI_IN_USE = False
        selectTreeItem(dlg)
    GUI_IN_USE = False

def closeRunWindow(progress, run_window):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        progress.child_window(auto_id="2", control_type="Button").click()
        run_window.child_window(title="취소", auto_id="2", control_type="Button").click()
    except:
        GUI_IN_USE = False
        closeRunWindow(progress, run_window)
    GUI_IN_USE = False
        
def continueProgress(progress):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        progress_msg = progress.child_window(auto_id="1263", control_type="Text").window_text()
        GUI_IN_USE = False
        while 'Execution complete.' not in progress_msg:
            Wait_GUI_IN_USE()
            GUI_IN_USE = True
            progress_msg = progress.child_window(auto_id="1263", control_type="Text").window_text()
            GUI_IN_USE = False
        return True
    except:
        GUI_IN_USE = False
        return False

def isMTTFailed(app, progress):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        app['FlamMap'].wait('ready', timeout=5)
        no_spread = app['FlamMap']
        msg = no_spread.child_window(title="No Spread From Ignition", auto_id="65535", control_type="Text").window_text()
        if msg == 'No Spread From Ignition':
            no_spread['확인'].click()
            GUI_IN_USE = False
            return False
    except:
        GUI_IN_USE = False
        return True

def launchMTT(run_window, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        run_window.child_window(title="Minimum Travel Time", control_type="TabItem").select()
        run_window.child_window(title="Launch MTT", auto_id="1066", control_type="Button").click()
        run_window.child_window(title="FlamMap Progress...", control_type="Window").wait('ready', timeout=timeout_)
        progress = run_window.child_window(title="FlamMap Progress...", control_type="Window")
    except:
        GUI_IN_USE = False
        progress = launchMTT(run_window, timeout_)
    GUI_IN_USE = False
    return progress

def get_run_window(dlg, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    run_window = dlg.child_window(title="Run : New Run", control_type="Window")
    try:
        run_window.wait('ready', timeout=timeout_)
    except:
        GUI_IN_USE = False
        import_run(dlg, timeout_)
        run_window = get_run_window(dlt, timeout_)
    GUI_IN_USE = False
    return run_window

def open_landscape(dlg, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        dlg.menu_select('Landscape -> Open Landscape')
        #dlg.menu_select("Landscape")
        #keyboard.send_keys('O')
        #keyboard.send_keys('{ENTER}')
        dlg.child_window(title="열기", control_type="Window").wait('ready', timeout=timeout_)
    except:
        GUI_IN_USE = False
        open_landscape(dlg, timeout_)
    GUI_IN_USE = False
        
def import_run(dlg, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    try:
        dlg.menu_select('Analysis Area -> Import Run')
        #dlg.menu_select("Analysis Area")
        #keyboard.send_keys('I')
        #keyboard.send_keys('{ENTER}')
        dlg.child_window(title="열기", control_type="Window").wait('ready', timeout=timeout_)
    except:
        GUI_IN_USE = False
        import_run(dlg, timeout_)
    GUI_IN_USE = False


def open_dialog(dlg, filepath, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    file_dlg = dlg.child_window(title="열기", control_type="Window")
    file_dlg.child_window(auto_id="1148", control_type="Edit").set_edit_text(filepath)
    file_dlg.child_window(title="열기(O)", control_type="Button").click()
    GUI_IN_USE = False

def save_dialog(dlg, filepath, timeout_):
    Wait_GUI_IN_USE()
    GUI_IN_USE = True
    dlg.child_window(title="다른 이름으로 저장", control_type="Window").wait('ready', timeout=timeout_)
    file_dlg = dlg.child_window(title="다른 이름으로 저장", control_type="Window")
    file_dlg.child_window(title="파일 이름:", auto_id="1001", control_type="Edit").set_edit_text(filepath)
    file_dlg.child_window(title="저장(S)", control_type="Button").click()
    time.sleep(2)
    try:
        dlg.child_window(title="다른 이름으로 저장", control_type="Window").wait('ready', timeout=5)
        GUI_IN_USE = False
        save_dialog(dlg, filepath, timeout_)
    except:
        pass
    GUI_IN_USE = False

def Wait_GUI_IN_USE():
    while GUI_IN_USE == True:
        time.sleep(0.5)

if __name__ == "__main__":
    print(DEVELOPER)
    searchPID()
    while True:
        pid = int(input("select PID(-1 is to run) : "))
        if pid < 0:
            break
        pids.append(pid)
    if len(pids) == 0:
        print("no PID was selected. Program is going to Terminate")
        exit()
    filepath = input("input Landscape Path : ")
    runpath = input("input Run Path : ")
    savepath = input("input Save Path : ")
    if filepath == "" or runpath == "" or savepath == "":
        print("path is invalid. Program is going to Terminate")
        exit()
    file_list = os.listdir(runpath)
    file_len = int(len(file_list)/2)
    batch = int(file_len/len(pids))
    start = 0
    end = start+batch
    global GUI_IN_USE
    GUI_IN_USE = False
    for pid_ in pids:
        print("start : ", start, " end : ", end)
        if start > file_len or end > file_len:
            break
        thr = threading.Thread(target=controlFlammap, args=(pid_, filepath,runpath,savepath, start+1, end, timeout_unit))
        threads.append(thr)
        thr.start()
        time.sleep(3)
        #controlFlammap(pid_, filepath,runpath,savepath, start, end)
        start = end
        end = start+batch
    for thrs in threads:
        thrs.join()
    
    print("Batch Process Finished")

    print(DEVELOPER)
```
